## 3.1 Spalte "successful_funding"

Zuerst wird untersucht, ob es auch Fälle gibt, bei denen zwar der funded_amount > 0 ist, aber das funding_goal nicht erreicht wurde. --> Auswertung unten zeigt, dass dies der Fall ist.

```python
df.insert(loc=3 , column='successful_funding', value='goal_not_achieved')
```

```python
df.loc[df.funded_amount == 0.0, 'successful_funding'] = 'no_funding'
#df
```

```python
df.loc[df.funded_amount == df.funding_goal, 'successful_funding'] = 'goal_achieved'
#df
```

```python
df.loc[df.funded_amount > df.funding_goal, 'successful_funding'] = 'goal_overachieved'
#df
```

## 3.3 & 3.4 Spalten "count_initiators" und "gender_initiators"

### Vorbereitung: Spalten "count_male" und "count_female"

```python
# Location ermitteln, nach dem die Spalte eingefügt werden soll
df.columns.get_loc('borrower_genders')
```

```python
# Liste erstellen aus der Spalte borrower_genders:
gender_list = df.borrower_genders.tolist()
```

```python
# zwei Spalten einfügen in df: eine für Frauen, eine für Männer
df.insert(16,'count_male' ,0,True)
df.insert(17,'count_female',0, True)
```

```python
df.info()
```

```python
# Datentypen nochmal auf integer ändern bevor wir die aufwändige Schleife rechnen:

df = df.astype({"use_description": int, "count_male": int, "count_female": int})
```

```python
df.info()
```

```python
# dieser Schritt wurde nicht erneut vor Export in HTML ausgeführt, da diese Berechnung mehrere Stunden benötigt. 
# Im nächsten Jupyter Notebook wird jedoch mit dem neuen Datensatz, welcher am Ende exportiert wurde, weiter gerechnet.
# leere temporäre Liste & Counter

temp_list = []
counter = 0

# for-Schleife
for i in gender_list:
    temp_list.clear()
    temp_list = str(i).split(', ')
    temp_male = temp_list.count('male') 
    temp_female = temp_list.count('female')
    df.iloc[counter,16] = temp_male
    df.iloc[counter,17] = temp_female
    counter = counter+1
```
